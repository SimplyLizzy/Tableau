```python
#Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time
import zipfile
import wget
import pandas as pd
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Scrape the website
url = "https://s3.amazonaws.com/tripdata/index.html"
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
data = []
for a in soup.find_all('a', href=True):
    data.append(a['href'])
data
```

```python
y2017 = []
y2018 = []
y2019 = []
y2020 = []
y2021 = []
y2022 = []

for i in range(0, len(data) -1):
    date = data[i].split('/')[-1].split('-')[0]
    
    if (date != "JC"):
        year = int(date[0:4])
        if (year == 2017):
            y2017.append(data[i])
        elif (year == 2018):
            y2018.append(data[i])
        elif (year == 2019):
            y2019.append(data[i])
        elif (year == 2020):
            y2020.append(data[i])
        elif (year == 2021):
            y2021.append(data[i])
        elif (year == 2022):
            y2022.append(data[i])

y2022
```

```python
dfile = y2022

for i in dfile:
    wget.download(i,"zip")
```

```python
zf = zipfile.ZipFile("zip/202203-citibike-tripdata.csv.zip") 
# if you want to see all files inside zip folder
zf.namelist()
```

```python
# now read your csv file 
df = pd.read_csv(zf.open("202203-citibike-tripdata.csv"))
df.to_csv('csv/202203-citibike-tripdata.csv') 

df.head()
```
